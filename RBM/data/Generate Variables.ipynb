```python
import numpy as np
import shutil
import os
from google.colab import drive
drive.mount('/content/drive')

part1 = np.round(np.random.uniform(-3.15, 3.15, size=(1000, 5)), 2)
part2 = np.round(np.random.uniform(-1.58, 1.58, size=(1000, 6)), 2)
Variables = np.hstack((part1, part2))

def count_identical_rows(matrix):
    # Find unique rows and their counts
    unique_rows, counts = np.unique(matrix, axis=0, return_counts=True)

    # Filter rows that appear more than once and return their counts
    duplicated_rows = unique_rows[counts > 1]
    duplicated_counts = counts[counts > 1]

    return dict(zip(map(tuple, duplicated_rows), duplicated_counts))

print(Variables)
# there is no reduplicate configuration.
print("reduplicate configurations:", count_identical_rows(Variables))
np.save('Variables.npy', Variables)
shutil.move('Variables.npy', '/content/drive/My Drive/Toric Code 2D/6: RBM on 3x3/Data/Variables.npy')
```
