```python
# This code generates 10000 random configurations
import numpy as np
import shutil
import os
from google.colab import drive
drive.mount('/content/drive')

np.random.seed(1)
S10000 = np.random.choice([-1, 1], size=(10000, 18))
print(S10000)


def count_identical_rows(matrix):
    # Find unique rows and their counts
    unique_rows, counts = np.unique(matrix, axis=0, return_counts=True)

    # Filter rows that appear more than once and return their counts
    duplicated_rows = unique_rows[counts > 1]
    duplicated_counts = counts[counts > 1]
    # Sum of all the duplicated counts
    total= sum(duplicated_counts)

    return {'total duplicated': total,
        'duplicated rows': dict(zip(map(tuple, duplicated_rows), duplicated_counts))}

# there is 358 reduplicate configurations.
print("reduplicate configurations:", count_identical_rows(S10000))
np.save('S10000.npy', S10000)
shutil.move('S10000.npy', '/content/drive/My Drive/Toric Code 2D/3: FRRBM on 3x3/Data/S10000.npy')
```
