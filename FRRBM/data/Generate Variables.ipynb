```python
import numpy as np
import shutil
import os
from google.colab import drive
drive.mount('/content/drive')

Variables9 = np.round(np.random.uniform(-3.15, 3.15, size=(1000, 9)), 2)

def count_identical_rows(matrix):
    # Find unique rows and their counts
    unique_rows, counts = np.unique(matrix, axis=0, return_counts=True)

    # Filter rows that appear more than once and return their counts
    duplicated_rows = unique_rows[counts > 1]
    duplicated_counts = counts[counts > 1]

    return dict(zip(map(tuple, duplicated_rows), duplicated_counts))

print(Variables9)
# there is no reduplicate configuration.
print("reduplicate configurations:", count_identical_rows(Variables9))
np.save('Variables9.npy', Variables9)
shutil.move('Variables9.npy', '/content/drive/My Drive/Toric Code 2D/3: FRRBM on 3x3/Data/Variables9_1k.npy')
```
